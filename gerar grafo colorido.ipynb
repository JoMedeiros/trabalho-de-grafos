```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from random import randint
```

```python
def ler_grafo(arquivo):
    arq = open(arquivo, 'r')
    texto = arq.readlines()
    G = nx.Graph()
    for linha in texto:
        if(linha[0] == 'c' or linha[0] == 'p'):
            continue
        linha = linha.replace("\n", "")
        partes = linha.split(" ")
        G.add_edge(partes[1], partes[2])

    arq.close()

    return G
```

```python
def let_coloracao(arquivo):
    arq = open(arquivo, 'r')
    texto = arq.readlines()
    C = dict();
    for linha in texto:
        if(linha == "\n"):
            return C
        
        linha = linha.replace("\n", "")
        partes = linha.split(" : ")
        C[partes[0]] = int(partes[1])

    arq.close()

    return C
```

```python
g = ler_grafo("Turmas/2018.2.col")
r = let_coloracao("2018.2.color")
```

```python
turmas = pd.DataFrame()
turmas['cor'] = pd.Series(index = turmas.index)
```

```python
co = ['#{0:0{1}x}'.format(randint(0,10000)*(int(i[0])), 6) for i in list(g.nodes.data())]
for i in range(0,len(r)):
    turmas.loc[i,['cor']] = co[r.get(list(g.nodes.data())[i][0])]
```

```python
nx.draw_kamada_kawai(g, node_color=turmas['cor'])
plt.draw()
```
